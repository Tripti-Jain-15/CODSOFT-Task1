```python
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

# Data Loading and Processing

```python
# Load the dataset
t_data = pd.read_csv(r'D:\Download\tested.csv',encoding = 'unicode_escape')
t_data
```

```python
# Info about data
t_data.info()
```

```python
t_data = t_data.drop(['PassengerId', 'Name', 'Ticket'], axis = 1)
t_data
```

```python
t_data['Sex'].replace({'male':0, 'female':1}, inplace = True)
t_data
```

```python
t_data['Embarked'].replace({'S':0, 'C':1, 'Q':2}, inplace = True)
t_data
```

# Handling Missing Values 

```python
# Checking number of null values in a column
t_data.isnull().sum()
```

```python
# Drop cabin column
t_data = t_data.drop(['Cabin'], axis = 1)
```

```python
t_data
```

```python
# Fill missing age values with the median age
t_data['Age'].fillna(t_data['Age'].mean(), inplace = True)
```

```python
# Fill missing fare values with the median fare
t_data['Fare'].fillna(t_data['Fare'].mean(), inplace=True)
```

```python
# Fill missing embarked values with the most common value
t_data['Embarked'].fillna(t_data['Embarked'].mode()[0], inplace=True)
```

```python
t_data.isnull().sum()
```

```python
t_data
```

# Splitting the Data

```python
# Split the data into input features (X) and target variable (Y)
X = t_data.drop(['Survived'], axis = 1)
Y = t_data['Survived']
```

```python
# Splitting the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=4)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
Y_test.shape
```

# Training Model

```python
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(X_train, Y_train)
```

# Model Evaluation 

```python
# Accuracy on training data
X_train_prediction = model.predict(X_test)
print(X_train_prediction)
```

```python
training_data_accuracy = accuracy_score(X_train_prediction , Y_test)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
print(X_test_prediction)
```

```python
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print(test_data_accuracy)
```
